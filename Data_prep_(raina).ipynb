```python
import pandas as pd
import numpy as np

raw = pd.read_csv("/content/data (1).csv")
raw.head()
```

```python
def clean_suicide(raw):
  df = pd.melt(raw, id_vars=['Countries, territories and areas'], var_name='Year', value_name='Suicide rate')
  df['Country'] = df['Countries, territories and areas'] + ', ' + df['Year']


  # drop the original 'Country' and 'Year' columns
  df = df.drop(['Countries, territories and areas', 'Year'], axis=1)

  # rearrange the columns as needed
  df = df[['Country', 'Suicide rate']]

  #Clean up suicide rates
  df['Suicide rate'] = df['Suicide rate'].apply(lambda x: x.split(' ')[0])
  return df

df = clean_suicide(raw)
df
```

```python
#Merge all happiness 

def clean_happiness(file, year):
  happy = pd.read_csv(file)
  happy = happy.drop(['Happiness Rank', 'Region', 'Dystopia Residual'], axis = 1)

  happy['Country'] = happy['Country'] + ',' + f' {year}'
  return happy

h1 = clean_happiness("/content/2015.csv", 2015)
h2 = clean_happiness("/content/2016.csv", 2016)
h3 = clean_happiness("/content/2017.csv", 2017)
h4 = clean_happiness("/content/2018.csv", 2018)
h5 = clean_happiness("/content/2019.csv", 2019)

happy = pd.concat([h1, h2, h3, h4, h5])

merged_df = pd.merge(df, happy, on='Country', how='left')
merged_df
```

```python
merged_brenda = pd.merge(merged_df, b, on='Country', how='left')
merged_brenda
```
